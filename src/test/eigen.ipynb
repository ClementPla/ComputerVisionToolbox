```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cov = 1

cov_mat = np.asarray([[1, cov], [-0.1, 1]])
data = np.random.multivariate_normal([0, 0], cov_mat, 1000)

plt.scatter(data[:, 0], data[:, 1])
plt.show()
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_mat@cov_mat.T)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```
